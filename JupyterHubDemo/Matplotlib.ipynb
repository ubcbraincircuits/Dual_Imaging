## A Pie Chart

```python
bars = ['Mahoneys', 'Gallery', 'Koerners', 'Porch']
ratings = [200,100,88,10]
plt.pie(ratings, labels=bars)
plt.title('Pie chart of `Favourite Bar on Campus` polled from 398 respondents')
```

### A Bar Chart

```python
pies = ['Apple', 'Cherry', 'Key Lime', 'Pumpkin']
ratings = [200,100,88,10]
plt.bar(x=pies, height=ratings)
plt.title('Bar chart of `Favourite Pies` polled from 398 respondents')
```
